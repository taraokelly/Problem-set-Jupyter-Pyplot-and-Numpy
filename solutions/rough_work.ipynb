```python
import numpy as np

sepal_length, sepal_width, petal_length, petal_width = np.genfromtxt('../data/IRIS.csv', delimiter=',', usecols=(0,1,2,3), unpack=True, dtype=float)
iris_class = np.genfromtxt('../data/IRIS.csv', delimiter=',', usecols=(4), unpack=True, dtype=str)
```

```python
import matplotlib.pyplot as plt

plt.scatter(sepal_length, sepal_width)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# Cluster size
c_size = 50

plt.plot(sepal_length[:c_size-1], sepal_width[:c_size-1],'r.', label='Iris Setosa')
plt.plot(sepal_length[c_size:c_size*2-1], sepal_width[c_size:c_size*2-1], 'g.', label='Iris Versicolor')
plt.plot(sepal_length[c_size*2:c_size*3-1], sepal_width[c_size*2:c_size*3-1],'b.', label='Iris Virginica')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```

```python
# Plot Setosa measurements 
plt.plot(petal_length, petal_width,'b.', label='Iris Setosa')

# https://github.com/emerging-technologies/emerging-technologies.github.io/blob/master/notebooks/simple-linear-regression.ipynb
# Calculate the best values for m and c.
petal_length_avg = np.mean(petal_length)
petal_width_avg = np.mean(petal_width)

petal_length_zero = petal_length - petal_length_avg
petal_width_zero = petal_width - petal_width_avg

m = np.sum(petal_length_zero * petal_width_zero) / np.sum(petal_length_zero * petal_length_zero)
c = petal_width_avg - m * petal_length_avg

# Plot best fit line 
plt.plot(petal_length, m * petal_length + c, 'forestgreen', label='Best fit line')
# Add labels
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend()
# Print graph
plt.show()
```

```python
# Calculate the R-squared value for our data set using numpy.
np.corrcoef(petal_length, petal_width)[0][1]**2
```
